```python
import turtle as t
t.bgcolor("light blue")
p = t.Turtle()
p.speed(6)
p.penup()
p.goto(150, 200)
p.pendown()
p.pensize(1)
p.fillcolor("yellow")
p.begin_fill()
p.circle(50)
p.end_fill()


i=1
while i< 5:
    p.penup()
    p.goto(150,250)
    p.rt(90)
    p.fd(50)
    p.pendown()
    p.fd(40)
    i+=1;

#nhánh nhỏ
p.penup()
p.goto(150,250)
p.rt(45)
p.fd(50)
p.pendown()
p.fd(25)
i=1
while i< 4:
    p.penup()
    p.goto(150,250)
    p.rt(90)
    p.fd(50)
    p.pendown()
    p.fd(25)
    i+=1;
    
p.penup()
p.goto(200,-200)
p.pendown()

p.fillcolor("brown")
p.begin_fill()
p.lt(45)
p.fd(60)
p.lt(90)
p.fd(30)
p.lt(90)
p.fd(60)
p.lt(90)
p.fd(30)
p.end_fill()


p.penup()
p.goto(185,-140)
p.pendown()
p.fillcolor("light green")
p.begin_fill()
p.fd(60)
p.lt(135)
p.fd(80)
p.lt(90)
p.fd(80)
p.lt(135)
p.fd(60)
p.end_fill()

p.penup()
p.goto(185,-85)
p.pendown()
p.fillcolor("light green")
p.begin_fill()
p.fd(60)
p.lt(135)
p.fd(80)
p.lt(90)
p.fd(80)
p.lt(135)
p.fd(60)
p.end_fill()

p.penup()
p.goto(185,-30)
p.pendown()
p.fillcolor("light green")
p.begin_fill()
p.fd(60)
p.lt(135)
p.fd(80)
p.lt(90)
p.fd(80)
p.lt(135)
p.fd(60)
p.end_fill()

p.penup()
p.goto(-300,-250)
p.pendown()
p.fillcolor("blue")
p.begin_fill()
p.fd(200)
p.lt(90)
p.fd(200)
p.lt(90)
p.fd(200)
p.lt(90)
p.fd(200)
p.end_fill()

p.penup()
p.lt(90)
p.fd(80)
p.pendown()
p.fillcolor("light green")
p.begin_fill()
p.fd(40)
p.lt(90)
p.fd(80)
p.lt(90)
p.fd(40)
p.lt(90)
p.fd(80)
p.end_fill()

p.penup()
p.goto(-300,-50)
p.lt(90)
p.pendown()
p.fillcolor("pink")
p.begin_fill()
p.fd(200)
p.lt(135)
p.fd(145)
p.lt(90)
p.fd(145)
p.end_fill()


p.penup()
p.bk(145)
p.pendown()
p.fillcolor("orange")
p.begin_fill()
p.bk(145)
p.lt(90)
p.fd(145)
p.rt(90)
p.fd(145)
p.end_fill()


p.penup()
p.fillcolor("yellow")
p.begin_fill()
p.lt(45)
p.fd(200)
p.pendown()
p.lt(120)
p.fd(120)
p.lt(60)
p.fd(250)
p.end_fill()



p.penup()

p.lt(135)
p.fd(100)
p.lt(45)
p.fd(80)
p.pendown()
p.fillcolor("dark red")
p.begin_fill()
p.fd(40)
p.lt(120)
p.fd(40)
p.lt(60)
p.fd(40)
p.lt(120)
p.fd(40)

p.end_fill()
t.done()









```

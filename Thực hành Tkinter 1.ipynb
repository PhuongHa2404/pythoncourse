### Bài 1:
Tạo một cửa sổ sẽ
yêu cầu người dùng nhập
tên. Khi họ nhấp vào
một nút nó sẽ hiển thị
 message “Hello” và
tên của họ và thay đổi
màu nền
và màu chữ của
messger box.

```python
import tkinter as tk

def display_message():
    name = entry.get()
    message = f"Hello {name}!"
    messagebox.config(text=message, fg="red", bg="yellow")

root = tk.Tk()
root.geometry("300x150")

label = tk.Label(root, text="Nhập tên:")
label.place(x=30, y=20)

entry = tk.Entry(root)
entry.place(x=100, y=20, width = 100, height=25)

button = tk.Button(root, text="Submit", command = display_message)
button.place(x=30, y=50, width= 50, height=25)

messagebox = tk.Label(root, text="")
messagebox.place(x=75, y=50, width=100, height=25)

root.mainloop()
```

### Bài 3
Tạo một chương trình sẽ hỏi
người dùng nhập một số trong một box.
 Khi họ bấm vào một button nó sẽ thêm số đó
thành tổng số và hiển thị nó trong một
button khác. Điều này có thể là
lặp đi lặp lại nhiều lần như
họ muốn và tiếp tục thêm vào
Tổng số. Nên có một button
khác thiết lập lại
tổng trở về 0 và làm trống
bản gốc text box , sẵn sàng cho
họ để bắt đầu lại.

```python
import tkinter as tk

class Summation:
    def __init__(self):
        self.total = 0
        
        # Tạo cửa sổ giao diện
        self.window = tk.Tk()
        self.window.title("Chương trình tính tổng")
        self.window.geometry("400x200")

        # Tạo box nhập liệu
        self.entry = tk.Entry(self.window, width=10, font=('Arial', 14))
        self.entry.pack(pady=10)

        # Tạo button tính tổng
        self.add_button = tk.Button(self.window, text="Tính tổng", font=('Arial', 14), command=self.add_number)
        self.add_button.pack()

        # Tạo button hiển thị tổng
        self.total_label = tk.Label(self.window, text=f"Tổng: {self.total}", font=('Arial', 14))
        self.total_label.pack(pady=10)

        # Tạo button reset
        self.reset_button = tk.Button(self.window, text="Reset", font=('Arial', 14), command=self.reset)
        self.reset_button.pack()

        self.window.mainloop()

    # Hàm tính tổng
    def add_number(self):
        try:
            number = int(self.entry.get())
            self.total += number
            self.total_label.config(text=f"Tổng: {self.total}")
            self.entry.delete(0, tk.END)
        except ValueError:
            self.entry.delete(0, tk.END)
            self.total_label.config(text="Lỗi: Vui lòng nhập số")

    # Hàm reset
    def reset(self):
        self.total = 0
        self.total_label.config(text=f"Tổng: {self.total}")
        self.entry.delete(0, tk.END)

# Chạy chương trình
app = Summation()
```

### Bài 4
Tạo một cửa sổ sẽ yêu cầu người dùng nhập một
tên trong một text box . Khi họ bấm vào một button nó
sẽ thêm nó vào cuối danh sách được hiển thị trên
màn hình. Tạo một button khác sẽ xóa
danh sách.

```python
import tkinter as tk

class Summation:
    def __init__(self):
        self.chuoi = ''
        
        # Tạo cửa sổ giao diện
        self.window = tk.Tk()
        self.window.title("Chương trình hiển thị chuỗi")
        self.window.geometry("400x200")

        # Tạo box nhập liệu
        self.entry = tk.Entry(self.window, width=10, font=('Arial', 14))
        self.entry.pack(pady=10)

        # Tạo button tính tổng
        self.add_button = tk.Button(self.window, text="Thêm", font=('Arial', 14), command=self.add_str)
        self.add_button.pack()

        # Tạo button reset
        self.reset_button = tk.Button(self.window, text="Xóa", font=('Arial', 14), command=self.reset)
        self.reset_button.pack()
        
        # Tạo button hiển thị tổng
        self.str_label = tk.Label(self.window, text=f"{self.chuoi}", font=('Arial', 14))
        self.str_label.pack(pady=10)
        self.window.mainloop()

    # Hàm tính tổng
    def add_str(self):
        self.chuoi += self.entry.get()
        self.str_label.config(text=f"{self.chuoi}")
        self.entry.delete(0, tk.END)

    # Hàm reset
    def reset(self):
        self.chuoi = ''
        self.str_label.config(text=f"{self.chuoi}")
        self.entry.delete(0, tk.END)

# Chạy chương trình
app = Summation()
```

Bài 5
1 km = 0,6214 dặm và 1dặm = 1,6093 km. sử dụng những con số này, làm cho một chương trìnhđiều đó sẽ cho phép người dùng chuyển đổi giữa dặm và km.

```python
import tkinter as tk

class ConvertValue:
    def __init__(self):
        self.giaTri = 0
        
        # Tạo cửa sổ giao diện
        self.window = tk.Tk()
        self.window.title("Chương trình quy đổi dặm <-> km")
        self.window.geometry("400x200")

        # Tạo box nhập liệu
        self.entry = tk.Entry(self.window, width=10, font=('Arial', 14))
        self.entry.pack(pady=10)

        # Tạo button tính tổng
        self.convertToKm = tk.Button(self.window, text="Dặm -> Km", font=('Arial', 14), command=self.convertToKm)
        self.convertToKm.pack()
        
        self.convertToMile = tk.Button(self.window, text="Km -> Dặm", font=('Arial', 14), command=self.convertToMile)
        self.convertToMile.pack()

        # Tạo button hiển thị tổng
        self.convert_label = tk.Label(self.window, text=f"", font=('Arial', 14))
        self.convert_label.pack(pady=10)

        self.window.mainloop()

    # Hàm tính tổng
    def convertToMile(self):
        try:
            #1 km = 0,6214 dặm và 1dặm = 1,6093 km
            number = float(self.entry.get())
            self.giaTri = number * 0.6214
            self.convert_label.config(text=f"{number} km = {self.giaTri} dặm")
        except ValueError:
            self.entry.delete(0, tk.END)
            self.convert_label.config(text="Lỗi: Vui lòng nhập số")

    def convertToKm(self):
        try:
            #1 km = 0,6214 dặm và 1dặm = 1,6093 km
            number = float(self.entry.get())
            self.giaTri = number * 1.6093
            self.convert_label.config(text=f"{number} dặm = {self.giaTri} km")
        except ValueError:
            self.entry.delete(0, tk.END)
            self.convert_label.config(text="Lỗi: Vui lòng nhập số")

# Chạy chương trình
app = ConvertValue()
```

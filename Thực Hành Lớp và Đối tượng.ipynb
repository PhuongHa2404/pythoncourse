Bài 1:

```python
""" Viết Class NhanVien gồm các thuộc tính :
+ Tên 
+ Tuổi 
+ Địa chỉ 
+ Tổng số giờ làm

Và các phương thức :
- Phương thức tạo
- input_info(): Nhập các thông tin cho nhân viên từ bàn phím
- print_info(): In ra tất cả các thông tin của nhân viên
- tinh_thuong(): Tính toán và trả về số tiền thưởng của nhân viên theo công thức sau :
Nếu tổng số giờ làm của nhân viên >= 200 thì thưởng = lương * 20%
Nếu tổng số giờ làm của nhân viên <200  và >= 100 thì thưởng = lương * 10%
Nếu tổng số giờ làm của nhân viên <100 thì thưởng = 0"""
```

```python
class NhanVien():
    def __init__ (self):
        self.ten = ''
        self.tuoi = 0
        self.diaChi = ''
        self.tongGioLam = 0
        self.luong = 0
    def input_info(self):
        self.ten = input("Nhập tên NV:")
        self.tuoi = int(input("Nhập tuổi NV:"))
        self.diaChi = input("Nhập địa chỉ:")
        self.tongGioLam = int(input("Nhập số giờ làm:"))
        self.luong = float(input("Nhập lương:"))
    def print_info(self):
        print(f'Thông tin nhân viên: - Tên:{self.ten}, Tuổi {self.tuoi}, Địa chỉ: {self.diaChi}, tổng giờ làm: {self.tongGioLam}')
    def tinh_thuong(self):
        thuong = 0
        if self.tongGioLam >= 200:
            thuong = self.luong * 20 / 100
        elif self.tongGioLam < 200 and self.tongGioLam >= 100:
            thuong = self.luong * 10 / 100
        else:
            thuong = 0
        print(f'Số tiền thưởng của bạn là: {thuong}')
nv = NhanVien()
nv.input_info()
nv.print_info()
nv.tinh_thuong()
```

```python
"""Tạo Class có thên SoHoc gồm có các thuộc tính và phương thức sau :
+ Thuộc tính : number1 ,number2
+ Phương thức :
- phương thức tạo __int()__
- input_info(): dùng để nhập 2 số number1 và number2
- print_info(): dùng để hiển number1 và number2
- addition(): dùng để cộng number1,number2
- subtract(): dùng để trừ number1 và number2
- multi(): dùng để nhân number1 và number2 
- division(): dùng để chia number1 và number2

Sau khi tạo class , khởi tạo 1 object SoHoc và in ra kết quả tính toán từ các hàm thực hiên phép tính"""
```

```python
class SoHoc():
    def __init__ (self):
        pass
    def input_info(self):
        self.number1 = float(input("Nhập số thứ nhất:"))
        self.number2 = float(input("Nhập số thứ hai:"))
    def print_info(self):
        print('Số thứ nhất là', self.number1)
        print('Số thứ hai là', self.number2)
    def addition(self):
        print(f'{self.number1 + self.number2}')
    def subtract(self): 
        print(f'{self.number1 - self.number2}')
    def multi(self): 
        print(f'{self.number1 * self.number2}')
    def division(self): 
        if self.number2 != 0:
            print(f'{self.number1 / self.number2}')
so = SoHoc()
so.input_info()
so.print_info()
so.addition()
so.subtract()
so.multi()
so.division()
```

```python
"""
Xây dựng lớp Phân số (Fraction) có các thành phần sau:
- Các thuộc tính: Tử số, mẫu số.
- Phương thức tạo để khởi tạo giá trị cho tử số và mẫu số.
- Các phương thức setter và getter.
- Các phương thức:
    + Nhập phân số
    + In Phân số
    + Rút gọn phân số
    + Nghịch đảo phân số
    + add(), sub(), mul(), div() tương ứng để thực hiện cộng, trừ, nhân, chia hai phân số cho nhau.
"""
```

```python
class PhanSo():
    def __init__(self, a, b):
        self.a = a
        self.b = b
        if b == 0:
            print('Mẫu số phải khác 0')
    
    def print_PS(self):
        return (f'{self.a}/{self.b}')
    def set_tuso(self, a):
        self.a = a
    def set_mauso(self, b):
        self.b = b
    
    def get_tuso(self):
        return self.a
    def get_mauso(self):
        return self.b
    def rut_gon(self):
        a = self.a
        b = self.b
        ucln = 1
        for i in range(1, min(abs(a), abs(b))+1):
            if i > ucln and a%i==0 and b%i==0:
                ucln=i
        a = a//ucln
        b = b//ucln
        #return PhanSo(a,b).print_PS()
    
    def nghich_dao(self):
        self.a, self.b = self.b, self.a
        return (f'{self.b}/{self.a}')
    def add(self, o):
        a = self.a*o.get_tuso() + self.b*o.get_mauso()
        b = self.b*o.get_mauso()
        return PhanSo(a,b).rut_gon()
            
    def sub(self, o):
        a = self.a*o.b - self.b*o.a
        b = self.b*o.b
        return PhanSo(a,b).rut_gon()
    
    def mul(self, o):
        a = self.a * o.a
        b = self.b * o.b
        return PhanSo(a,b).rut_gon()
    
    def div(self, o):
        a = self.a * o.b
        b = self.b * o.a
        return PhanSo(a,b).rut_gon()
    
phanSo = PhanSo(5,9)
phanSo2 = PhanSo(6,9)
print(f'{phanSo.add(phanSo2)}')
print(f'{phanSo.sub(phanSo2)}')
print(f'{phanSo.mul(phanSo2)}')
print(f'{phanSo.div(phanSo2)}')
```
